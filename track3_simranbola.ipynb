```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
athelets_df= pd.read_csv('https://raw.githubusercontent.com/rgriff23/Olympic_history/master/data/athlete_events.csv')
athelets_df
```

```python
athelets_df.info()
```

```python
athelets_df.describe()
```

```python
df_noc= pd.read_csv('https://raw.githubusercontent.com/rgriff23/Olympic_history/master/data/noc_regions.csv')
df_noc
```

```python
df_noc.info()
```

```python
df_noc= df_noc.drop(['notes'], axis=1)
df_noc
```

```python
final_df= pd.merge(athelets_df, df_noc, how='left', on='NOC')
final_df
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
final_df.isnull().sum()
```

Filing the missing values:


```python
final_df['Age'].fillna(value=final_df['Age'].mean(),inplace=True)
```

```python
final_df['Height'].fillna(value=final_df['Height'].mean(),inplace=True)
```

```python
final_df['Weight'].fillna(value=final_df['Weight'].mean(),inplace=True)
```

```python
final_df['region'].fillna(value="Region Unknown",inplace=True)
```

```python
final_df['Medal'].fillna(value="No Medal",inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.duplicated()
```

```python
final_df.duplicated().sum()
```

```python
final_df.drop_duplicates(keep='first',inplace=True)
```

#Exploratory Data Analysis

Medal distribution by country

```python
fig_df= final_df[final_df['Medal'].notna()].groupby('NOC')['Medal'].count().sort_values(ascending=False).reset_index().head(10)
fig= px.bar(fig_df, barmode='stack', color= 'Medal' ,title='Medal Distribution by Country')
fig.show()
```

Top 10 Gold,Silver and Bronze Winning Nations

```python
medal_df=final_df[final_df['Medal']=="Gold"].groupby(["Team"]).count().sort_values(by='ID', ascending=False).reset_index()
medal_df.head(10)
```

```python
gold_plot = px.pie(medal_df.head(10), values='ID',hole=0.6, names='Team', title='<b>Top 10 Gold Winning Nations</b>',color='Team',template = "none")
fig.update_layout(plot_bgcolor = " whitesmoke")
gold_plot.show()
```

```python
silver_df=final_df[final_df['Medal']=="Silver"].groupby(["Team"]).count().sort_values(by='ID', ascending=False).reset_index()
silver_df.head(10)
```

```python
silver_plot = px.pie(silver_df.head(10), values='ID',hole=0.6, names='Team', title='<b>Top 10 Silver Winning Nations</b>',color='Team',template = "none")
fig.update_layout(plot_bgcolor = " whitesmoke")
silver_plot.show()
```

```python
bronze_df=final_df[final_df['Medal']=="Bronze"].groupby(["Team"]).count().sort_values(by='ID', ascending=False).reset_index()
bronze_df.head(10)
```

```python
bronze_plot = px.pie(bronze_df.head(10), values='ID',hole=0.6, names='Team', title='<b>Top 10 Bronze Winning Nations</b>',color='Team',template = "none")
fig.update_layout(plot_bgcolor = " whitesmoke")
bronze_plot.show()
```
